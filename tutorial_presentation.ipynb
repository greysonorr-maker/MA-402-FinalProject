# PETSc Least-Squares Solver (Augmented System)

This notebook demonstrates the petsc4py implementation of solving a least-squares problem using the augmented system method.

## Problem

We want to solve:

minimize ||Ax - b||_2

using the augmented system:

[ -I A ] [ r ] = [ b ]
[ A^T 0 ] [ x ] [ 0 ]

```python
from tutorial_module import solve_augmented_least_squares
import numpy as np

A = np.array([
    [1,1,0],
    [1,0,1],
    [0,1,1],
    [1,1,1]
])

b = np.array([2,3,3,4])

x, r, res = solve_augmented_least_squares(A, b)

print("x =", x)
print("residual =", res)
```

## Results

The solver successfully computes the least-squares solution.

From the output, the solution is:

x = [1, 1, 2]

The residual norm is very close to zero, which shows the solution is accurate.

```python
import matplotlib.pyplot as plt

x = [1, 1, 2]

plt.bar(["x0", "x1", "x2"], x)
plt.title("Least-Squares Solution")
plt.show()
```

## Conclusion

This notebook demonstrates how PETSc can solve least-squares problems using the augmented system method.

Instead of using the normal equations, this approach is more numerically stable and allows us to solve for both the solution and residual at the same time.
